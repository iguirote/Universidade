```python
# Simulador de Combate: Turnos de RPG Em um jogo de turno, um herói tem 100 de HP e enfrenta um monstro com 100 de HP.
#O programa deve pedir ao usuário o dano do ataque do herói e o dano do ataque do monstro por rodada.
# Use um while que continue enquanto ambos estiverem com HP acima de zero. Ao final de cada turno, mostre o HP restante de ambos.
#Se alguém chegar a 0, encerre e anuncie o vencedor.

heroi = 100
monstro = 100

while heroi > 0 and monstro > 0:
    ataque_heroi = int(input("Digite o dano do ataque do herói: "))
    ataque_monstro = int(input("Digite o dano do ataque do monstro: "))
    heroi = heroi - ataque_monstro
    monstro = monstro - ataque_heroi
    print("O herói ficou com ",  heroi, " de vida")
    print("O monstro ficou com ", monstro, " de vida")
print("Fim de jogo!")
if heroi > 0:
    print("O herói venceu!")
else:
    print("O monstro venceu!")
```

```python
#Monitor de Inundação (Sentinela)
#Você está desenvolvendo um sistema de alerta para uma Smart City. O programa deve ler o nível do rio (em metros) continuamente.
#* Se o nível for menor que 3m: "Estado Normal".
#* Entre 3m e 5m: "Estado de Alerta".
#* Acima de 5m: "Evacuação Imediata".
#O programa só deve encerrar a leitura se o usuário digitar um valor negativo (flag/bandeira de encerramento).

rio = 0
print("Digite n° negativo para encerrar!")
while rio >= 0:
    rio = float(input("Digite o nível do rio: "))
    if rio < 0:
        break
    if rio < 3:
        print("Estado Normal")
    if rio >= 3 and rio <= 5:
        print("Estado de Alerta")
    if rio > 5:
        print("Evacuação Imediata")
print("Fim do programa")
```

```python
#Média de Temperatura Global para um painel informativo sobre mudanças climáticas, você precisa calcular a média de temperatura de um período.
#O usuário deve informar quantos dias deseja analisar. Use o while para ler a temperatura de cada um desses dias.
#Ao final, o programa deve exibir a média aritmética das temperaturas e informar se a média está "Acima do esperado" (caso seja maior que 25°C) ou "Dentro da normalidade".

dias = int(input("Digite a quantidade de dias: "))
tempDias = dias
temperatura = 0
while dias > 0:
    temperatura = temperatura + float(input("Digite a temperatura do dia: "))
    dias = dias - 1
media = temperatura / tempDias
if media > 25:
    print("A média de",media,"está acima do esperado")
else :
    print("A média de",media,"está dentro da normalidade")
```
