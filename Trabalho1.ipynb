```python
# 1. Criação do Arquivo de Log Temporário
log_content = """Aragorn;Guerreiro;15;2;4500
Legolas;Arqueiro;22;1;5200
Gandalf;Mago;10;5;8000
Frodo;Ladrão;1;10;500
Gimli;Guerreiro;12;4;3800
Boromir;Guerreiro;8;8;2500
Galadriel;Mago;5;0;9500
Eowyn;Guerreiro;11;3;4100
Faramir;Arqueiro;9;4;3200
Saruman;Mago;18;2;8800
Sauron;Mago;30;1;15000
Gollum;Ladrão;2;15;300
Sam;Ladrão;4;2;1200
Merry;Ladrão;3;4;800
Pippin;Ladrão;2;5;700
Elrond;Mago;7;1;6000
Arwen;Mago;4;0;4500
Theoden;Guerreiro;10;6;3500
Ugluk;Guerreiro;6;12;2000
Lurtz;Arqueiro;14;5;4800"""

with open("partida.txt", "w", encoding="utf-8") as f:
    f.write(log_content)
```

```python
# 2. Leitura e Tratamento de Dados
def ler_jogadores(caminho):
    jogadores = []

    with open(caminho, 'r', encoding='utf-8') as arquivo:
        for linha in arquivo:
            linha = linha.strip()

            if linha == "":
                continue

            dados = linha.split(";")

            jogador = {
                "nome": dados[0],
                "classe": dados[1],
                "kills": int(dados[2]),
                "deaths": int(dados[3]),
                "dano": int(dados[4])
            }

            jogadores.append(jogador)

    return jogadores
```

```python
# 3. Cálculo do KDA
def calculo_kda(kills, deaths):
    if deaths == 0:
        return kills
    else:
        return kills / deaths
```

```python
# 4. Filtro de Classe
def filtro_classe(jogadores, classe):
    jogadores_classe = []

    for jogador in jogadores:
        if jogador["classe"].lower() == classe.lower():
            jogadores_classe.append(jogador)

    return jogadores_classe
```

```python
# 5. Cálculo dos Destaques da Partida
def destaques(jogadores):
    print("\n--- DESTAQUES DA PARTIDA ---")

    maior_dano = 0
    nome_maior_dano = ""

    for jogador in jogadores:
        if jogador["dano"] > maior_dano:
            maior_dano = jogador["dano"]
            nome_maior_dano = jogador["nome"]

    print(f"Maior dano: {nome_maior_dano} - {maior_dano}")

    total_kills = 0

    for jogador in jogadores:
        total_kills += jogador["kills"]

    media_kills = total_kills / len(jogadores)
    print(f"Média de kills: {media_kills:.2f}")

    print("\nJogadores com KDA superior a 2.0:")
    for jogador in jogadores:
        kda = calculo_kda(jogador["kills"], jogador["deaths"])
        if kda > 2.0:
            print(jogador["nome"].upper())
```

```python
# 6. Execução Principal do Programa
def main():
    caminho = "partida.txt"
    todos_jogadores = ler_jogadores(caminho)
    # Troca a classe aqui para mostrar apenas aquela classe
    classe = "Guerreiro"

    lista = filtro_classe(todos_jogadores, classe)
    print(f"--- {classe.upper()} ---")

    for j in lista:
        print(j["nome"])

    destaques(todos_jogadores)


if __name__ == "__main__":
    main()
```
